```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x_vals = np.linspace(0, 1.4, 1001)

uni_lo = .3
uni_hi = 1.1
norm_lo = .7
norm_hi = 1.3
norm_mean = 1.0
norm_std = 1.0

uniform_dist = stats.uniform(loc=uni_lo, scale=(uni_hi-uni_lo))
normal_dist = stats.truncnorm((norm_lo-norm_mean)/norm_std, (norm_hi-norm_mean)/norm_std, loc=norm_mean, scale=norm_std)
```

```python
plt.figure(figsize=(6,6))
plt.plot(x_vals, normal_dist.pdf(x_vals), 'k--', label='relative demand multiplier')
plt.plot(x_vals, uniform_dist.pdf(x_vals), 'r-', label='total demand multiplier')

plt.grid(True)
plt.xlabel('x, -')
plt.ylabel('probability density of x, -')
plt.xlim([0, 1.4])
plt.ylim([0, 1.8])
plt.legend()
```
